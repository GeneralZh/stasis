```python
import graphlab as gl
```

```python
sts_train = gl.SFrame.read_csv('sts.csv', delimiter='\t', column_type_hints=[str, str, float, str, str], quote_char='\0')
train_asiya = gl.SFrame.read_csv('MegaEXPERT/sts_asiya.csv', delimiter='\t', quote_char='\0')
train_beer_reval = gl.SFrame.read_csv('MegaEXPERT/sts_beer_reval.csv', delimiter='\t', quote_char='\0')
train_DLS2015 = gl.SFrame.read_csv('MegaEXPERT/sts_DLS2015.csv', delimiter='\t', quote_char='\0')
train_glove = gl.SFrame.read_csv('MegaEXPERT/sts_glove.csv', delimiter='\t', quote_char='\0')
train_meteor = gl.SFrame.read_csv('MegaEXPERT/sts_meteor.csv', delimiter='\t', quote_char='\0')
train_DLS2014 = gl.SFrame.read_csv('MegaEXPERT/sts_DLS2014.csv', delimiter='\t', quote_char='\0')
```

```python
assert len(sts_train) == len(train_asiya) == \
len(train_beer_reval) == len(train_DLS2015) == \
len(train_glove) == len(train_meteor) == len(train_DLS2014)
print len(sts_train)
```

```python
sts_train.add_columns(train_asiya)
sts_train.add_columns(train_beer_reval)
sts_train.add_columns(train_DLS2015)
sts_train.add_columns(train_DLS2014)
sts_train.add_columns(train_glove)
sts_train.add_columns(train_meteor)
```

```python
sts_train.column_names()
```

```python
sts_test = gl.SFrame.read_csv('sts2016-annotated/sts2016-test.clean.csv', delimiter='\t', column_type_hints=[str, str, str, str], quote_char='\0')
test_asiya = gl.SFrame.read_csv('sts2016-annotated/sts2016-test.asiya.csv', delimiter='\t', quote_char='\0')
test_beer_reval = gl.SFrame.read_csv('sts2016-annotated/sts2016-test.beer_reval.csv', delimiter='\t', quote_char='\0')
test_DLS2015 = gl.SFrame.read_csv('sts2016-annotated/sts2016-test.DLS_compose.csv', delimiter='\t', quote_char='\0')
test_glove = gl.SFrame.read_csv('sts2016-annotated/sts2016-test.glove.csv', delimiter='\t', quote_char='\0')
test_meteor = gl.SFrame.read_csv('sts2016-annotated/sts2016-test.meteor.csv', delimiter='\t', quote_char='\0')
test_DLS2014 = gl.SFrame.read_csv('sts2016-annotated/sts2016-test.DLS2014.csv', delimiter='\t', quote_char='\0')
```

```python
assert len(sts_test) == len(test_asiya) == \
len(test_beer_reval) == len(test_DLS2015) == \
len(test_glove) == len(test_meteor) == len(test_DLS2014)
print len(sts_test)
```

```python
sts_test.add_columns(test_asiya)
sts_test.add_columns(test_beer_reval)
sts_test.add_columns(test_DLS2015)
sts_test.add_columns(test_DLS2014)
sts_test.add_columns(test_glove)
sts_test.add_columns(test_meteor)
```
