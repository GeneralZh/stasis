```python
sts_train = sts_train.dropna(columns=['Score'])
len(sts_train)
```

```python
import graphlab
import numpy


train = sts_train

feat = [i for i in train.column_names() if i not in ['Dataset', 'Domain', 'Score', 'Sent1', 'Sent2']]

train_valid_shuffled = graphlab.toolkits.cross_validation.shuffle(train, random_seed=1)

def kth_start_end(n,k,kth):
    start = (n*kth)/k
    end = (n*(kth+1))/k-1
    return start, end+1

def get_kth_train(data, kth_start, kth_end, n):
    part1 = data[0:kth_start]
    part2 = data[kth_end+1:n]
    return part1.append(part2)

n = len(train)
accuracy = []
k = 10
```

```python
for i in range(k):
    start, end = kth_start_end(n,k,i)
    valid_kth = train_valid_shuffled[start:end]
    train_kth = get_kth_train(train_valid_shuffled, start, end, n)
    m = graphlab.random_forest_regression.create(train_kth, target='Score', features=feat, validation_set=valid_kth)
```

```python
for i in range(k):
    start, end = kth_start_end(n,k,i)
    valid_kth = train_valid_shuffled[start:end]
    train_kth = get_kth_train(train_valid_shuffled, start, end, n)
    m = graphlab.boosted_trees_regression.create(train_kth, target='Score', features=feat, validation_set=valid_kth)    
```

```python
for i in range(k):
    start, end = kth_start_end(n,k,i)
    valid_kth = train_valid_shuffled[start:end]
    train_kth = get_kth_train(train_valid_shuffled, start, end, n)
    m = graphlab.linear_regression.create(train_kth, target='Score', features=feat, validation_set=valid_kth)
```
